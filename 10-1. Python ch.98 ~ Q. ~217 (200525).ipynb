### ■ 표본평균의 표본분포
#### 모집단으로부터 추출한 표본의 평균을 표본평균이라 하며, 기호로는 모집단 평균을 나타내는 $\mu$와 구분하여 $\bar{x}$로 표시한다.
#### 하나의 모집단으로부터 표본의 크기($n$)을 바꿔가며 많은 종류의 표본을 추출할 수 있으며, 또한 같은 크기($n$)의 서로 다른 표본을 추출할 수 있는 경우의 수도 매우 많다.

    점심시간 문제 : 모집단은 초등학생 100만 명
    표본의 개수는 100명 씩 추출하여 100명의 평균값을 구한다
    표본의 평균값을 10만 개 모은다

### ※(점심시간 문제)(파이썬 알고리즘 문제34번을 다시 하나씩 이해하기 위한 문제)

2000 년 초등학교 5학년 어린이들의 전국 평균키는 148.5cm , 표준편차는 7.8 이였습니다.

식생활의 변화로 인해 어린이들의 성장에 변화가 일어났다고 생각됩니다. 그래서 현대의 초등학교 5학년 어린이들을 100명을 무작위로 추출해 평균키를 구하는 작업을 100000 번 하여 표본의 평균키 십만개를 a 라는 비어있는 리스트에 넣으시오 ~  

````python
import numpy as np

avg=148.5
std=7.8
N=1_000_000
height=np.random.randn(N)*std+avg
a=[np.random.choice(height,100).mean() for _ in range(int(N/10))]
a
````
