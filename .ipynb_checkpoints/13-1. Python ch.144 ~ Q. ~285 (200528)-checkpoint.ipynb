````python
def chu_scr(n):

    import urllib.request 
    from bs4 import BeautifulSoup 

    params1=[]
    for i in range(1,n+1):
    
        list_url=f"https://news.joins.com/Search/JoongangNews?page={i}\
&Keyword=%EC%BD%94%EB%A1%9C%EB%82%98&SortType=New&SearchCategoryType=JoongangNews"

        url=urllib.request.Request(list_url) 
        rs=urllib.request.urlopen(url).read().decode("utf-8") 
        soup=BeautifulSoup(rs, "html.parser")

        rs1=soup.find_all('h2',class_='headline mg')
        for j in rs1:
            for k in j:
                params1.append(k.get("href"))
    return params1

list_url=chu_scr(1)
print(list_url)
````

### ※ 문제285. 코로나 관련한 기사들을 전부 스크롤링 할 수 있도록 아래의 url을 첫번째 함수에 넣어서 for loop 문으로 페이지 번호 3개만 스크롤링 할 수 있게 하시오

````python
def chu_scr(n):

    import urllib.request 
    from bs4 import BeautifulSoup 

    params1=[]
    for i in range(1,n+1):
    
        list_url=f"https://news.joins.com/Search/JoongangNews?page={i}\
&Keyword=%EC%BD%94%EB%A1%9C%EB%82%98&SortType=New&SearchCategoryType=JoongangNews"

        url=urllib.request.Request(list_url) 
        rs=urllib.request.urlopen(url).read().decode("utf-8") 
        soup=BeautifulSoup(rs, "html.parser")

        rs1=soup.find_all('h2',class_='headline mg')
        for j in rs1:
            for k in j:
                params1.append(k.get("href"))
    return params1

def chu_detail_scr(n):

    import urllib.request 
    from bs4 import BeautifulSoup 
    
    list_url=chu_scr(n)
    rs2=''
    f=open('k:/Itwill/2. Python/자료/mytext20.txt','w',encoding='utf-8')
    for i in list_url:
        url=urllib.request.Request(i) 
        rs=urllib.request.urlopen(url).read().decode("utf-8") 
        soup=BeautifulSoup(rs, "html.parser")
        rs1=soup.find_all('div',id='article_body')
        
        for j in rs1:
            rs2 += j.get_text(" ",strip=True) +'\n'
    return rs2
       
print(chu_detail_scr(3))
````

### ※(점심시간 문제) 백분위수 양측 5% 점 상한값과 하한값이 출력되게 하시오
```
2000년 초등학교 5학년 어린이들의 전국 평균키는 148.5cm , 분산은 7.8의 제곱 였다. 
식생활등의 변화로 인해 어린이들의 성장에 변화가 일어났다고 생각된다. 
그래서 현대의 초등학교 5학년 아이들을 무작위로 100명을 추출했다. 그 평균키는 149.2cm 였다. 
초등학교 5학년 아이들의 키에 변화가 일어났는지를 유의수준 5%로 검정해 보자.
다만 분산은 변함이 없다고 가정한다. 
```

결과:  백분위수 양측 5%점 상한값과 하한값이 출력되게 하시오 !

````python
def tall_hypo(n):
    import random
    import numpy as np
    from scipy import stats
        
    avg=148.5 # 모평균
    std=7.8 # 모표준편차
    N=1_000_000 # 모집단 데이터 수

    height=np.random.randn(100_000)*std+avg # 10만개 샘플링
    a=[np.random.choice(height,100).mean() for i in range(int(N/10))] # 표본평균분포
    
    s_std=np.std(a) # 표본표준편차
    s_avg=149.2 # 샘플평균(고정수치)
    
    rs1=avg+1.96*s_std # 95% 신뢰구간의 상한 임계값 : 표본평균+1.96*std
    rs2=avg-1.96*s_std # 9%% 신뢰구간의 하한 임계값 : 표본평균-1.96*std
    
    return f'상한:{round(rs1,2)}, 하한:{round(rs2,2)}'

print(tall_hypo(5))    
````

````python
def tall_hypo(n):
    import random
    import numpy as np
    import scipy
        
    avg=148.5 # 모평균
    std=7.8 # 모표준편차
    N=1_000_000 # 모집단 데이터 수
    
    confidence=1-(n/100) # 신뢰구간 설정

    height=np.random.randn(100_000)*std+avg # 10만개 샘플링
    a=[np.random.choice(height,100).mean() for i in range(int(N/10))] # 표본평균분포
    
    df = len(a) # 표본의 개수
    s_avg, se = 149.2, np.std(a) # 평균 149.2, 표본표준편차
    h = abs(round(se * scipy.stats.t.ppf((1 - confidence) / 2., df-1),2)) # 표준정규분포 상 임계값 계산, df-1 : 자유도 설정
    return f'하한:{round(s_avg-h,2)}, 상한:{round(s_avg+h,2)}'
    
print(tall_hypo(5))
print(tall_hypo(1))
````

````python
from statistics import NormalDist
def confidence_interval(data, confidence=0.95):
    dist = NormalDist.from_samples(data)
    z = NormalDist().inv_cdf((1 + confidence) / 2.)
    h = dist.stdev * z / ((len(data) - 1) ** .5)
    return dist.mean - h, dist.mean + h
````
