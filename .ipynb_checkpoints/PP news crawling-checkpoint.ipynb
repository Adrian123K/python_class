```python
def head_scr(n):

    import urllib.request 
    from bs4 import BeautifulSoup 

    params1=[]
    title=''
    dueday=['070'+str(i) for i in range(1,n+1)]
    h=open('headline.txt','w',encoding='utf-8')
    print(dueday)
    for day in dueday :
        list_url=f"https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=105&date=2020{day}"
        print(list_url)
        url=urllib.request.Request(list_url) 
        rs=urllib.request.urlopen(url).read().decode("cp949") 
        soup=BeautifulSoup(rs, "html.parser")

        rs1=soup.find_all('a', class_='nclicks(rnk.sci)')

        for j in rs1:
            title += j.get_text('title',strip=True)+'\n'
            params1.append(j.get_text('title',strip=True))            
        head = [i for i in params1 if i]
    h.write(title)
    h.close()
    return head

head_scr(3)
```

```python
def head_scr(n):

    import urllib.request 
    from bs4 import BeautifulSoup 

    params1=[]
    dueday=['070'+str(i) for i in range(1,n)]
    
    for day in dueday :
        list_url=f"https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=105&date=2020{day}"
        
        for i in range(1,n+1):


            url=urllib.request.Request(list_url) 
            rs=urllib.request.urlopen(url).read().decode("cp949") 
            soup=BeautifulSoup(rs, "html.parser")

            rs1=soup.find_all('a',class_='nclicks(rnk.sci)')
            
            for j in rs1:
                for k in j:
                    params1.append(k.get("href"))
        return params1

list_url=head_scr(10)
print(list_url)
```
