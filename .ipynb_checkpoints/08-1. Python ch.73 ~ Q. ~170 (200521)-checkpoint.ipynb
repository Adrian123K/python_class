### ※ Pandas에서의 merge
```python
문법: 
    pandas.merge(emp_left, dept_right, how='inner',on='deptno')
                     m   :    1
옵션:
    how='inner'  : 열의 데이터가 양쪽 데이터 프레임에 공통적으로 존재하는 교집합일 경우에만 추출
    how='outer'  : 열의 데이터가 양쪽 데이터 프레임에 공통으로 존재하는 교집합이 아니어도 추출
    how='left'   : 왼쪽 데이터 프레임의 키 열에 속하는 데이터 값을 기준으로 병합
    how='right'  : 오른쪽 데이터 프레임의 키 열에 속하는 데이터 값을 기준으로 병합
```

````python
import pandas as pd

emp=pd.read_csv('k:/Itwill/2. Python/자료/emp3.csv')
dept=pd.read_csv('k:/Itwill/2. Python/자료/dept3.csv')
result = pd.merge(emp, dept, on='deptno', how='left')
result
````

### ※ 문제168. 부서번호, 부서번호별 평균월급을 출력하시오

```sql
select deptno, avg(sal)
    from emp
    group by deptno;
    ```

### ※ 문제169. 사원 테이블에 avgsal라는 파생변수를 추가하고 자기가 속한 부서번호의 평균월급으로 값을 갱신하시오

````python
from my_loc import oracle as oc
import pandas as pd

emp=oc.connect('emp')
rs=emp[['DEPTNO','SAL']]
rs1=rs.groupby('DEPTNO')['SAL'].mean().reset_index().astype(int)
rs1.columns.values[1]='avgsal'
rs2=pd.merge(emp, rs1, on='DEPTNO',how='left')
rs2
````

### ※ 문제170.(점심시간 문제) 퇴사할 것으로 예상되는 직원을 예측하기 위해서 under_sal이라는 column을 emp에 추가하고 자기가 속한 부서번호의 평균월급보다 자기의 월급이 작거나 같으면 1로 갱신하고 아니면 0으로 갱신

````python
rs2['under_sal']=(rs2['SAL']<=rs2['avgsal']).astype(int)
rs2
````

### ※ (점심시간 문제)(파이썬 알고리즘 문제 30번) 초등학생의 평균키가 160라는 주장이 신뢰구간 95%안에 드는지 아닌지 알아내시오
#### 파이썬 알고리즘 22번 참고하여 모집단 구성
```python
print(child_tall(160))
신뢰구간 95% 안에 있지 않습니다
print(child_tall(142))
신뢰구간 95% 안에 있습니다.
```

````python
def child_tall(n):
    import numpy as np
    from scipy.stats import norm
    import matplotlib.pyplot as plt

    x=np.arange(120,160,0.001)
    y=norm.pdf(x,140,5) # 평균 140에 표준편차 5에 해당하는 확률밀도함수
    plt.plot(x,y,color='red')
    plt.axvline(x=n, color='y', linestyle='-', linewidth=3)
    plt.fill_between(x, y, where=abs(140-x)<1.96*5, interpolate=True, color='green', alpha=0.5)
        
    if abs(140-n)<1.96*5:
        return '신뢰구간 95% 안에 있습니다.'
    else:
        return '신뢰구간 95% 안에 있지 않습니다.'
    
print(child_tall(160))   
print(child_tall(142))
````
