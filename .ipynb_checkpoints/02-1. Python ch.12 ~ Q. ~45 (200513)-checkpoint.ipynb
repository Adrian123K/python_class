### ■ 예제13. for 문 개념 배우기 Ⅲ (for ~ else)
#### for 반복문을 완전히 수행했을 때만 실행하는 부분을 정의하려면 for ~else 문을 사용  
for ~ else 에서 else 뒤에서 실행되는 코드는 for 반복문을 성공적으로 수행해야지만 실행

````python
# 예제
for i in range(1,11):
    print(i) # 반복 실행 코드
else:
    print('Perfect') # for문이 모두 실행되었을 때 실행할 코드
````

#### ※ 설명 : for 반복문이 break에 의해 중단없이 정상적으로 모두 실행되어야만 특정 코드를 실행하게 할 경우 for ~ else 를 사용

### ※ 문제22. 아래의 예제 코드에 break를 넣어서 1부터 10까지 출력할 때 물어보는 숫자에서 멈추게 하고 
###                 전체가 다 돌면 perfect이 아래에 메세지로 나오게 하시오
숫자를 입력하세요 ~ 7  

````python
a=int(input('숫자를 입력하세요 ~ '))
for i in range(1,11):
    print(i)
    if i==10:
        continue
    elif i==a:
        break
else:
    print('Perfect')
````

### ※ (파이썬 알고리즘 문제 12번) 6개의 제품이 들어있는 상자가 있는데 그 중 2개가 불량품이라고 하자. 
### 제품 검사를 위해서 3개를 추출했을 때 적어도 1개가 불량품일 확률은?
#### 수학식으로 하지 않고 파이썬으로 상자 안에 아래와 같이 제품의 6개를 넣고 랜덤으로 추출해서 수행
box=['정상','정상','불량','정상','불량','정상']

````python
import random

box=['정상','정상','불량','정상','불량','정상']
ls=[]
cnt=0
for i in range(100000):
    a, b, c=random.choice(box),random.choice(box),random.choice(box)
    ls.append(a)
    ls.append(b)
    ls.append(c)
    if '불량' in ls: 
        cnt += 1
    ls.clear()
print(cnt/100000)
````

````python
ls=[]
for i in range(3):
    a, b, c=random.choice(box),random.choice(box),random.choice(box)
    ls.append(a)
    ls.append(b)
    ls.append(c)
    print(ls)
    if len(ls)==3:
        ls.clear()
        print(ls)
````

````python
import random

box=['정상','정상','불량','정상','불량','정상']
cnt=0
for i in range(100000):
    a=random.choice(box)
    b=random.choice(box)
    c=random.choice(box)
    if '불량' in (a, b, c):
        cnt +=1
print(cnt/100000)  
````

````python
import copy

box=['정상','정상','불량','정상','불량','정상']
box1=copy.deepcopy(box)
cnt=0
for i in range(100000):
    a=random.choice(box1)
    box1.remove(a)
    b=random.choice(box1)
    box1.remove(b)
    c=random.choice(box1)
    box1.remove(c)
    if '불량' in (a, b, c):
        cnt +=1
        box1=copy.deepcopy(box)
print(cnt/100000)
````
